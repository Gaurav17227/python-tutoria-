```python
# find the largest number in list

number = [12,20,34,65,78,54,34,21]
largest = number[0]
print("given list:" , number)

for i in number:
  if i > largest:
    largest = i
print("largest number in list:" , largest)
```

```python
# remove the dublicate number from the list using loop

number = [1,2,3,3,3,4,5,6,5,7,4,7,5,8]
print("original list:", number)

unique = []

for i in number :
  if i not in unique:
    unique.append(i)

print("unique list:" , unique)
```

# RECURSION

```python
def factorial(n):
  if n == 1:
    return 1
  return n * factorial(n-1)

print("factorial of 7 : " , factorial(7))
```
